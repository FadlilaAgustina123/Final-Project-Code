```python
import seaborn as sns
df = df.drop(['name'], axis=1)
plt.figure(figsize=(17,17))
sns.heatmap(df.corr(), vmax=1, vmin=-1, center=0, cmap="Spectral", annot=True)
```

```python
# seleksi variabel dengan tipe data object
categorical_var = df.select_dtypes("object")

drop_var_cat = []
cat_var = []
for kol in categorical_var:
    num_cat = len(df[kol].unique())
    if num_cat > 100:
        drop_var_cat.append(kol)
    else:
        cat_var.append(kol)
        
print("##########--Data Kategorik--################")
print(cat_var)
print("##########--TOO MUCH VARIABLE--################")
print(drop_var_cat)
```

```python
# merubah data kategori menjadi numerik
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
label_encoder = LabelEncoder()
encoded_data = df[cat_var].apply(label_encoder.fit_transform)
encoded_data
```

```python
# pilih kolom
list_columns_full = df.columns
list_encoded = encoded_data.columns

list_columns = []
for col in list_columns_full:
  if col not in list_encoded:
    list_columns.append(col)

print(list_columns)
```

```python
# menggabungkan kolom
cluster_data = pd.concat([encoded_data, df[list_columns]], axis=1)
cluster_data.head()
```

```python
# melihat persebaran informasi instagram dan facebook
prices = ['instagram', 'facebook']
for col in prices:
    plt.figure()
    sns.boxplot(x=col, data=df)
    plt.tight_layout()
```

```python
# melihat persebaran informasi traffic dan employees
prices = ['traffic', 'employees']
for col in prices:
    plt.figure()
    sns.boxplot(x=col, data=df)
    plt.tight_layout()
```

```python
# Data Scaling
list_columns = cluster_data.columns

scaler = StandardScaler()
transformation_data = scaler.fit(cluster_data[list_columns])
cluster_data[list_columns] = transformation_data.transform(cluster_data[list_columns])
cluster_data.head()
```

```python
# perulangan ini digunakan untuk mengeksekusi nilai k dengan data
# selanjutnya akan dicari SSE

Sum_of_squared_error = []
K = range(2,10)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(cluster_data)
    Sum_of_squared_error.append(km.inertia_)
```

```python
plt.plot(K, Sum_of_squared_error, 'bx-')
plt.xlabel('k')
plt.ylabel('SSE')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# memodelkan data dengan k-means
cluster = KMeans(n_clusters=3)
cluster.fit(cluster_data)
```

```python
# membuat kolom baru berdasarkan hasil clustering
segment = pd.Series(cluster.labels_, name="segment")
```

```python
# menggabungkan data dengan hasil cluster
results = pd.concat([cluster_data, segment], axis=1)
results.head()
```

```python
# Pengelompokan traffic berdasarkan instagram dan twitter
sns.scatterplot(data=results, x='instagram', y='twitter', hue='segment', palette="deep")
```

```python
fig, axs = plt.subplots(figsize=(17,9), ncols=2)
fig.suptitle("Perbandingan Traffic Instagram dan Facebook Tiap Cluster")

# kolom 1 baris 1
sns.boxplot(x='segment', y='instagram', data = results, ax=axs[0]).set_title("Traffic Facebook Tiap Cluster")

# kolom 2 baris 1
sns.boxplot(x='segment', y='facebook', data = results, ax=axs[1]).set_title("Traffic Instagram")

plt.show()
```

```python
# kolom 2 baris 1
sns.boxplot(x='segment', y='key', data = results).set_title("Kunci")
plt.show()
```
